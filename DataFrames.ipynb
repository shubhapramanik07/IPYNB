```python
import numpy as np
import pandas as pd
```

```python
data = {'Name': ['John', 'Anna', 'Peter', 'Linda'],
    'Age': [28, 34, 29, 42],
    'City': ['New York', 'Paris', 'Berlin', 'London'],
    'Salary': [65000, 70000, 62000, 85000]
}
df = pd.DataFrame(data)
```

```python
df
```

```python
data_list = [
    ['John', 28, 'New York', 65000],
    ['Anna', 34, 'Paris', 70000],
    ['Peter', 29, 'Berlin', 62000],
    ['Linda', 42, 'London', 85000]
]
df2 = pd.DataFrame(data_list)
df2
# when you are using a list instead of dict, you will find the name of the column as 0,1,2,3 like this...
```

```python
# we want to add specific names instead of 0,1,2,3
cols = ["name","age","city","salary"]
df2 = pd.DataFrame(data_list,columns = cols)
df2
```

```python
df2
```

```python
df2["name"]
# single brackets when you want to select a single column
```

```python
df2[["name","city"]]
# double brackets when you want to select multiple columns
```

CREATING NEW COLUMN:

```python
df2["designation"] = ["engg","dr","teacher","pilot"]
df2
# adding a new column to the dataframe
```

REMOVING COLUMNS:

```python
df2.drop("designation",axis=1)
# this drops the column 'designation' from the dataframe
# where axis=0 is for rows and axis=1 is for columns
```

```python
# if we only use drop, then it will never be removed permanently, but to do so, we have to use inplace = True
df2.drop("designation",axis=1,inplace = True)
df2
```

```python
df2.drop(["name","city"],axis=1)
# this drops multiple columns but does not do it permanently as inplace is not used
```

```python
# removing rows:--------------------
df2.drop(0,axis=0)
```

SELECTING ROWS:

```python
df2.loc[0]
# this will give the first row of the dataframe
```

```python
df2.loc[[1,2]]
# this will give multiple rows at once and the output will be a dataframe btw we are using double brackets here.
```

SELECTING SUBSETS OF ROWS & COLUMNS:

```python
df[["Age","Salary"]].loc[[0,3]]
# first we selected multiple columns and then multiple rows using double brackets for both selections
```

```python
df2.loc[[0,1]][["city","salary"]]
# another way of doing the same thing as above
```

CONDITIONAL SELECTION:

```python
df2
```

```python
# i only want to see those people whose age is above 30
df2[df2["age"] > 30]
```

```python
# i only want to see people whose age is 30  and their city must be in paris:
df2[(df2["age"]>30 ) & (df2["city"] == "Paris")]
```
