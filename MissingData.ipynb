```python
import numpy as np
import pandas as pd
```

```python
data = {
    'A' : [1,2,np.nan,4,5],
    'B' : [1,2,3,4,5],
    'C': [1, 2, 3, np.nan, np.nan],
    'D': [1, np.nan, np.nan, np.nan, 5]
}
df = pd.DataFrame(data)
df
```

FINDING MISSING DATA:

```python
df.isna().sum() 
# this shows how many null numbers are in each column
```

```python
df.isna().any()
# this tells us there is a null value present in every column
```

REMOVING MISSING DATA:

```python
df.dropna()
# this drops any row with a null value
```

```python
df.dropna(thresh=3)
# it doesn't remove the rows where you must have 3 or more non-null values, but does its vice-versa
```

FILLING THE MISSING DATA:

```python
df.fillna(0)
# replace the nan vale with 0
```

```python
values = {'A':0,'B':100,'C':377,'D':699}
df.fillna(value = values)
# we can customise to put the different values in the replacement of nan 
```

```python
df.fillna(df.mean())
# we can also replace nan with the overall mean value
```
