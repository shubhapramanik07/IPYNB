```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# years are given with sales data for 4 products (A, B, C, D)  for 1 year to 4 year column-wise

sales_data = np.array([
    [1, 200, 220, 250, 270],  # Sales data for Product A
    [2, 150, 180, 200, 230],  # Sales data for Product B
    [3, 300, 320, 350, 400],   # Sales data for Product C
    [4, 100, 120, 130, 150]   # Sales data for Product D
])

print("++++  Tomato Sales Analysis ++++")
print("\n Sales data shape: ",sales_data.shape)
print("\n Sample data for 1st 3 restaurant: ",sales_data[0:3])

# print(sales_data[:,1:])

```

```python

#? total sales per year

print(np.sum(sales_data, axis = 0))

yearly_total = np.sum(sales_data[:,1:], axis = 0)
print("\n Yearly total sales: ",yearly_total)

#? minimum sales per restaurant

min_sales = np.min(sales_data[:,1:], axis = 1)
print("\n Minimum sales in a year for each product: ",min_sales)

#? maximum sales per year
max_sales = np.max(sales_data[:,1:], axis = 0)
print("\n Maximum sales in a year for each product: ",max_sales)

#? average sales per restaurant
avg_sales = np.mean(sales_data[:,1:], axis = 1)
print("\n Average sales in a year for each product: ",avg_sales)

```

```python
# cumulative sales calculation
cumulative_sales = np.cumsum(sales_data[:,1:], axis = 1)

print("\n Cumulative sales for each product over the years:\n",cumulative_sales)

# to print or plot cumulative sales for each product over the years using matplotlib

plt.figure(figsize=(10, 6))
# Plot cumulative sales for each product
# years = ['Year 1', 'Year 2', 'Year 3', 'Year 4']
# products = ['Product A', 'Product B', 'Product C', 'Product D']

plt.title('Cumulative Sales for Each Product Over Years')
plt.plot(np.mean(cumulative_sales, axis=0), marker='o')
plt.xlabel('Years')
plt.ylabel('Cumulative Sales')
plt.grid(True)
plt.show()
```

## starting vector...


```python
v1 = np.array([10, 20, 30])
v2 = np.array([5, 15, 25])

print("vector addition: ",v1 + v2)

vector_subtraction = v1 - v2

print("vector subtraction: ", vector_subtraction)

vector_multiplication = v1 * v2
print("vector multiplication: ", vector_multiplication)

vector_division = v1 / v2
print("vector division: ", vector_division)

print("Dot product: ", np.dot(v1, v2))

angle = np.arccos(np.dot(v1,v2) / (np.linalg.norm(v1) * np.linalg.norm(v2)))
print("Angle between v1 and v2 (in radians): ", angle)
print("Angle between v1 and v2 (in degrees): ", np.degrees(angle))
```
