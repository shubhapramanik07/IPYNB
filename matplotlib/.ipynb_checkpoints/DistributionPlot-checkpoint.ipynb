```python
import seaborn as sns
import matplotlib.pyplot as plt
```


**We use distribution plots to visualize the distribution of quantitative data.**

Let's discuss some plots that allow us to visualize the distribution of a data set. These plots are:

***1.distplot/histplot
2. jointplot
3. pairplot
4. rugplot***

```python
df = sns.load_dataset('tips')
df
# If we go to "https://github.com/mwaskom/seaborn-data" GitHub repo, 
# then we will find a lot of csv files over there and we can use of it.
```

```python
df['size'].unique()
# We can determine whether the size column is quantitative or qualitative!!!
```

// As we are getting a fixed element over the whole column, it's a qualitative...

```python
plt.subplot(2,1,1)
# as we know how to create a subplot in the previous file (row,col,pos)
sns.histplot(df['total_bill'],bins = 20, kde = True)
# this is how we can create a histplot or distplot// using bins, we can instruct how many columns should be there //
# kde=True adds a smooth curve on top of the histogram that shows the probability density (the shape of the distribution).

plt.subplot(2,1,2)
sns.histplot(df['tip'],bins = 20,kde = True)
```

we use sns(seaborn). Here we don't need any specific things to do; we just have to give the variable name, which graph we are preparing for, and some minimal info ..

```python
# this is how we join two coordinates of two different quantitative values...
sns.jointplot(x='total_bill',y = 'tip',data = df,kind='reg')
# default ( kind = 'scatter' ) but here I write regression
```

```python
# we can pairplot where we just have to give the whole dataset and a hue value
# and if want various color then write after hue (palette='rainbow')
sns.pairplot(df,hue='size')
```

```python
# this is the useless...we are not gonna use it anymore ig...
# btw A rug is a thick cloth or carpet used on the floor for decoration or comfort. and 
# the thing in the picture is looking like the same ... right ??
sns.rugplot(df['tip'])
```
