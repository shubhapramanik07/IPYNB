```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

<!-- Matplotlib Basics -->

```python
x = np.linspace(0,5,11)
x
y = x **3 
y
```

```python
plt.plot(x,y)
plt.title("AIML Journey")
plt.xlabel("time")
plt.ylabel("journey")
```

<!-- Subplots in Matplotlib -->

```python
plt.subplot(2,2,1)
# here 2 = width, 2 = height, 1 means the 1st box(sub-box) of a big box
plt.plot(x,y)
plt.subplot(2,2,2)
# here 2 = width, 2 = height, 2 means the 2st box(sub-box) of a big box
plt.plot(y,x)
plt.subplot(2,2,4)
plt.plot(x**4,y**5)
plt.subplot(2,2,3)
plt.plot(x**2,y**9)
```

**Object Oriented way**

Matplotlib allows the aspect ratio, DPI and figure size to be specified when the Figure object is created. You can use the figsize and dpi keyword arguments.

figsize is a tuple of the width and height of the figure in inches dpi is the dots-per-inch (pixel per inch).

```python
fig = plt.figure(figsize=(20,8),dpi=111)

axis1 = fig.add_axes([0.1,0.1,0.4,0.8])
axis1.plot(x,y)
axis2 = fig.add_axes([0.3,0.4,.3,.3])
axis2.plot(x,y)
```
